# Example

```python
# Load images
im1 = imread(r"images\sbug00.png")
im2 = imread(r"images\sbug01.png")
im3 = imread(r"images\sbug02.png")
im4 = imread(r"images\sbug03.png")
im5 = imread(r"images\sbug04.png")
im6 = imread(r"images\sbug05.png")
im7 = imread(r"images\sbug06.png")
im8 = imread(r"images\sbug07.png")
im9 = imread(r"images\sbug08.png")
im10 = imread(r"images\sbug09.png")
im11 = imread(r"images\sbug10.png")
im12 = imread(r"images\sbug11.png")
im13 = imread(r"images\sbug12.png")
```

```python
# Each image is blurred in a place and clear in the rest
plt.rcParams['figure.figsize'] = (18, 18)
plt.subplot(4,2,1)
plt.imshow(im1)
plt.subplot(4,2,2)
plt.imshow(im2)
plt.subplot(4,2,3)
plt.imshow(im3)
plt.subplot(4,2,4)
plt.imshow(im4)
plt.subplot(4,2,5)
plt.imshow(im5)
plt.subplot(4,2,6)
plt.imshow(im6)
plt.subplot(4,2,7)
plt.imshow(im7)
plt.subplot(4,2,8)
plt.imshow(im8)
```

```python
# Image fusion
# It takes a few seconds
a1=fusion(im1,im2)
a2=fusion(a1,im3)
a3=fusion(a2,im4)
a4=fusion(a3,im5)
a5=fusion(a4,im6)
a6=fusion(a5,im7)
a7=fusion(a6,im8)
a8=fusion(a7,im9)
a9=fusion(a8,im10)
a10=fusion(a9,im11)
a11=fusion(a10,im12)
a12p=fusion(a11,im13)
```

```python
# Final image
plt.rcParams['figure.figsize'] = (14, 8)
imshow(a12p)
```
